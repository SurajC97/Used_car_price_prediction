# CAR PRICE PREDICTION

### Problem statement

With the covid 19 impact in the market, we have seen lot of changes in the car market. Now some
cars are in demand hence making them costly and some are not in demand hence cheaper. One
of our clients works with small traders, who sell used cars. With the change in market due to covid
19 impact, our client is facing problems with their previous car price valuation machine learning
models. So, they are looking for new machine learning models from new data. We have to make
car price valuation model. This project contains two phase

In this phase, we have to scrape at least 5000 used cars data. we need to scrape the data of used cars from websites (Olx, cardekho, Cars24
etc.) we need web scraping for this. we have to fetch data for different locations and  include all types of cars in our data for example- SUV, Sedans, Coupe, minivan,
Hatchback.

```python
# importing libraries

import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
# initializing the driver to a variable

driver= webdriver.Chrome('chromedriver.exe')
```

```python
# Giving address to the driver

driver.get('https://www.cardekho.com/')
```

```python
# Clicking on the used cars option.

driver.find_element_by_xpath("//nav[@class='gsc_col-sm-12 gsc_col-md-12']/ul/li[2]/span/span").click()
time.sleep(2)

# Fetching the city page's urls.

cities=[]
city= driver.find_elements_by_xpath("//li[@class='gsc_col-xs-4 gsc_col-sm-3 gsc_col-md-2']/a")
for i in city:
    cities.append(i.get_attribute('href'))
```

```python
# Creating lists to store data.

year=[]
comp=[]
model=[]
var=[]
fuel=[]
kms=[]
onrs=[]
trans=[]
milage=[]
seat=[]
loc=[]
price=[]
cars=[]

# Scraping the required data.

for i in cities:
    driver.get(i)
    time.sleep(3)
    
    # Extracting the urls of the car pages.
    
    driver.execute_script("window.scrollTo(0, 4000)")
    time.sleep(2)
    driver.execute_script("window.scrollTo(4000, 6000)")
    time.sleep(2)
    driver.execute_script("window.scrollTo(6000, 8000)")
    time.sleep(2)
    driver.execute_script("window.scrollTo(8000, 10000)")
    time.sleep(3)
    driver.execute_script("window.scrollTo(10000, 12000)")
    time.sleep(3)
    driver.execute_script("window.scrollTo(12000, 14000)")
    time.sleep(3)
    driver.execute_script("window.scrollTo(14000, 16000)")
    time.sleep(3)
    driver.execute_script("window.scrollTo(16000, 18000)")
    time.sleep(5)
    car= driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7 carsName']/a")
    for j in car:
        cars.append(j.get_attribute('href'))
        
    # Opening the car pages and scraping the data.
    for k in cars:
        driver.get(k)
        time.sleep(2)
        
        # Scraping year of manufacture.
        try:
            y= driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li/div/div")
            year.append(y.text)
        except NoSuchElementException:
            y= driver.find_element_by_xpath("//ul[@class='gsc_row detailsList']/li/div/div[2]")
            year.append(y.text)
            
        # Scraping company name.   
        try:
            c= driver.find_element_by_xpath("//div[@class='vdp-head']/h1[2]")
            comp.append(c.text.split(' ')[0]) 
        except NoSuchElementException:
            c= driver.find_element_by_xpath("//div[@class='paddingBorder clearfix']/div/h1")
            comp.append(c.text.split(' ')[1])
            
        # Scraping model name.
        try:
            m= driver.find_element_by_xpath("//div[@class='vdp-head']/h1[2]")
            model.append(m.text.split(' ')[1]) 
        except NoSuchElementException:
            m= driver.find_element_by_xpath("//div[@class='paddingBorder clearfix']/div/h1")
            model.append(m.text.split(' ',2)[-1])
            
        # Scraping the variant.
        try:
            v= driver.find_element_by_xpath("//div[@class='vdp-head']/h1[2]")
            var.append(v.text.split(' ',2)[-1]) 
        except NoSuchElementException:
            v= driver.find_element_by_xpath("//div[@class='paddingBorder clearfix']/div/div[1]")
            var.append(v.text)
    
        # Scraping feul type.
        try:
            f= driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[5]/div/div")
            fuel.append(f.text)
        except NoSuchElementException:
            f= driver.find_element_by_xpath("//ul[@class='gsc_row detailsList']/li[3]/div/div[2]")
            fuel.append(f.text)
        
        # Scraping kilometers ran.
        try:
            l= driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[3]/div/div")
            kms.append(l.text)
        except NoSuchElementException:
            l= driver.find_element_by_xpath("//ul[@class='gsc_row detailsList']/li[4]/div/div[2]")
            kms.append(l.text)
            
        # Scraping number of owners.
        try:
            o= driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[4]/div/div")
            onrs.append(o.text)
        except NoSuchElementException:
            o= driver.find_element_by_xpath("//ul[@class='gsc_row detailsList']/li[6]/div/div[2]")
            onrs.append(o.text)
            
        # Scraping transmition type.
        try:
            t= driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[6]/div/div")
            trans.append(t.text)
        except NoSuchElementException:
            t= driver.find_element_by_xpath("//ul[@class='gsc_row detailsList']/li[8]/div/div[2]")
            trans.append(t.text)
            
        # Scraping mileage.
        try:
            m= driver.find_element_by_xpath("//div[@class='SpecsFeatureList']/ul/li/div[2]")
            milage.append(m.text)
        except NoSuchElementException:
            m= driver.find_element_by_xpath("//div[@id='specification-wdght']/div/div/div/ul/li[1]/div/div[2]")
            milage.append(m.text)
            
        # Scraping number of seats.
        
        try:
            s= driver.find_element_by_xpath("//div[contains(text(),'Seats')]/../div[2]")
            seat.append(s.text)
        except NoSuchElementException:
            seat.append('-')
            
        # Scraping location.
        
        loc.append(i.split('+')[-1])
            
        # Scraping the price.
        try:
            p= driver.find_element_by_xpath("//span[@class='amount']")
            price.append(p.text)
        except NoSuchElementException:
            p= driver.find_element_by_xpath("//div[@class='priceSection']/span[2]")
            price.append(p.text) 
            
    # emptying the cars list.
    cars=[]
```

```python
# Creating dataframe for the data.

cars= pd.DataFrame({"Year of manufacture":year,"Company":comp,"Model":model,"Variant":var,"Fuel type":fuel,"Location":loc,
                     "Kilometers driven":kms,"No. of Owners":onrs,"Transmission":trans,"Mileage":milage,"Seats":seat,"Price":price})
cars
```

```python
# Giving another address to the driver

driver.get('https://www.cars24.com/')
```

```python
# Clicking on view all cars

v= driver.find_element_by_xpath("//span[@class='jVs6G']/a")
all_cars= v.get_attribute('href')
v.click()
```

```python
# Creating lists to store data.

year=[]
comp=[]
model=[]
var=[]
fuel=[]
kms=[]
onrs=[]
trans=[]
milage=[]
seat=[]
loc=[]
price=[]
cars=[]

# Scraping the required data.

for i in range(1,5):
    
    driver.get(all_cars)
    time.sleep(2)
    driver.find_element_by_xpath("//p[@class='_1OR7d']").click()  # Clicking on city option
    time.sleep(2)
    driver.find_element_by_xpath("//ul[@class='_16Bvy']/li[{}]".format(i)).click() # Selecting the city
    time.sleep(3)
    
    # Extracting the urls of the car pages.
    driver.execute_script("window.scrollTo(0,5000)")
    time.sleep(6)
    car= driver.find_elements_by_xpath("//div[@class='col-4']/a")
    for j in car:
        cars.append(j.get_attribute('href'))
        
    # Opening the car pages and scraping the data.
    for r in cars:
        driver.get(r)
        time.sleep(1)
        
        # Scraping year of manufacture.
        y= driver.find_element_by_xpath("//h2[@class='_2geSF']")
        year.append(y.text.split(' ')[0])
            
        # Scraping company name.   
        c= driver.find_element_by_xpath("//h2[@class='_2geSF']")
        comp.append(c.text.split(' ')[1])
            
        # Scraping model, variant, transmission type.
        m= driver.find_element_by_xpath("//h2[@class='_2geSF']")
        v= driver.find_element_by_xpath("//span[@class='_18vo2']")
        t= driver.find_element_by_xpath("//span[@class='_18vo2']/small")
        
        model.append(m.text.split(' ',2)[2].replace(v.text,''))
        var.append(v.text.replace(t.text,''))
        trans.append(t.text)
    
        # Scraping feul type.
        f= driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[3]")
        fuel.append(f.text)
        
        # Scraping kilometers ran.
        k= driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[1]")
        kms.append(k.text)
            
        # Scraping number of owners.
        o= driver.find_element_by_xpath("//ul[@class='_2tp8h']/li[5]")
        onrs.append(o.text)
        
        # Scraping location.
        l= driver.find_element_by_xpath("//p[@class='_2NHUv']/span")
        loc.append(l.text)
            
        # Scraping the price.
        p= driver.find_element_by_xpath("//strong[@class='_2yYvS']")
        price.append(p.text.split('₹')[-1])
        
        try:
            driver.find_element_by_xpath("//ul[@class='Qqw8W']/li[3]").click() # clicking on the specification option.
            time.sleep(1)
        except NoSuchElementException:
            pass
        
        # Scraping mileage.
        try:
            g= driver.find_element_by_xpath("//label[contains(text(),'Mileage (kmpl)')]/../strong")
            milage.append(g.text)
        except NoSuchElementException:
            milage.append('-')
            
        # Scraping number of seats.
        try:
            s= driver.find_element_by_xpath("//label[contains(text(),'Seating capacity')]/../strong")
            seat.append(s.text)
        except NoSuchElementException:
            seat.append('-')
            
    # emptying the cars list.
    cars=[]
```

```python
# Creating dataframe for the data.

cars1= pd.DataFrame({"Year of manufacture":year,"Company":comp,"Model":model,"Variant":var,"Fuel type":fuel,"Location":loc,
                    "Kilometers driven":kms,"No. of Owners":onrs,"Transmission":trans,"Mileage":milage,"Seats":seat,"Price":price})
cars1
```

```python
# Finding element for search bar and entering city name into it.

bar= driver.find_element_by_id("placesQuery")
bar.send_keys(cities[0])
time.sleep(2)
driver.find_element_by_id("placesQuery").send_keys(Keys.ENTER)
```

```python
# Getting the address to used car home page.

used_cars=[]
u= driver.find_element_by_xpath("//div[@class='grid-4 omega font15 text-black']/div/ul/li/a")
used_cars=u.get_attribute('href')
```

```python
# Creating lists to store data.

year=[]
comp=[]
model=[]
var=[]
fuel=[]
kms=[]
onrs=[]
trans=[]
milage=[]
seat=[]
loc=[]
price=[]
cars=[]

# Scraping the required data.

for i in range(0,10):
    
    driver.get(used_cars)
    driver.find_element_by_id("global-place").click()  # Clicking on city option
    time.sleep(2)
    
    # Entering the city name.
    bar= driver.find_element_by_id("placesQuery")
    bar.send_keys(cities[i])
    time.sleep(2)
    driver.find_element_by_id("placesQuery").send_keys(Keys.ENTER)
    time.sleep(3)
    
    # Extracting the urls of the car pages.
    driver.execute_script("window.scrollTo(0,9000)")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0,-1)")
    time.sleep(1)
    driver.execute_script("window.scrollTo(0,25000)")
    driver.execute_script("window.scrollTo(0,35000)")
    time.sleep(10)
    car= driver.find_elements_by_id("linkToDetails")
    for j in car:
        cars.append(j.get_attribute('href'))
        
    # Opening the car pages and scraping the data.
    for r in cars:
        driver.get(r)
        time.sleep(1)
        
        # Scraping year of manufacture.
        y= driver.find_element_by_id('content-title')
        year.append(y.text.split(' ')[0])
            
        # Scraping company name.   
        c= driver.find_element_by_xpath("//div[@class='breadcrumb margin-bottom15']/ul/li[4]/a/span")
        comp.append(c.text)
            
        # Scraping model name.
        m= driver.find_element_by_xpath("//div[@class='breadcrumb margin-bottom15']/ul/li[5]/a/span")
        model.append(m.text)

        # Scraping variant.
        v= driver.find_element_by_id("car-breadcrumb-link")
        var.append(v.text)

        #Scraping transmission type.
        t= driver.find_element_by_xpath("//ul[@class='basic-details']/li[4]/span[2]")
        trans.append(t.text)

        # Scraping feul type.
        f= driver.find_element_by_xpath("//ul[@class='basic-details']/li[3]/span[2]")
        fuel.append(f.text.replace(' + Cng',''))
        
        # Scraping kilometers ran.
        k= driver.find_element_by_xpath("//ul[@class='basic-details']/li[2]/span[2]")
        kms.append(k.text)
            
        # Scraping number of owners.
        o= driver.find_element_by_xpath("//div[contains(text(),'No. of Owner(s)')]/../div[2]")
        onrs.append(o.text)
        
        # Scraping location.
        loc.append(cities[i])
            
        # Scraping the price.
        p= driver.find_element_by_xpath("//p[@class='used-actual']")
        price.append(p.text.replace('₹ ',''))
        
        # clicking on the specification option.
        driver.find_element_by_xpath("//div[@class='cw-tabs cw-tabs-flex cw-tabs-inner-margin0']/ul/li[3]").click()
        
        # Scraping mileage.
        try:
            g= driver.find_element_by_xpath("//div[@class='equal-width'][contains(text(),'kmpl')]")
            milage.append(g.text)
        except NoSuchElementException:
            try:
                g= driver.find_element_by_xpath("//div[@class='equal-width'][contains(text(),'km/kg')]")
                milage.append(g.text)
            except NoSuchElementException:
                milage.append('-')
            
        # Scraping number of seats.
        try:
            s= driver.find_element_by_xpath("//div[@class='equal-width'][contains(text(),'Person')]")
            seat.append(s.text)
        except NoSuchElementException:
            seat.append('-')
            
    # emptying the cars list.
    cars=[]
```

```python
# Creating dataframe for the data.

cars= pd.DataFrame({"Year of manufacture":year,"Company":comp,"Model":model,"Variant":var,"Fuel type":fuel,"Location":loc,
                    "Kilometers driven":kms,"No. of Owners":onrs,"Transmission":trans,"Mileage":milage,"Seats":seat,"Price":price})
cars4
```
